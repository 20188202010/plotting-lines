```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([[1,1],[1,2],[2,1],[2,2],[1.5,1.5],
              [2.5,10],[2.5,11],[3.5,10],[3.5,11],[3,10.5],
              [8,5],[9,6],[8,6],[9,5],[8.5,5.5],
              [4,5],[5,6],[4,6],[5,5],[4.5,5.5],
             ])
```

```python
colorize = dict(c=X[:, 0], cmap=plt.cm.get_cmap('rainbow', 4))

# Plot the data
fig = plt.figure()
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.grid()

plt.axis('equal');
```

```python
from sklearn.cluster import KMeans
```

```python
def do_kmeans(init_centers,X):
    kmeans = KMeans(n_clusters=2,init=init_centres,algorithm='full')
    kmeans.fit(X)
    labels = kmeans.predict(X)
    print(labels)
    centroids = kmeans.cluster_centers_
    print(centroids)
    colmap = {1: 'r', 2: 'g', 3:'y'}
    colors = list(map(lambda x: colmap[x+1], labels))
    plt.scatter(X[:,0],X[:,1], color=colors, alpha=0.5, edgecolor='k')
    plt.grid()
    for idx, centroid in enumerate(centroids):
        plt.scatter(*centroid, color=colmap[idx+1])

    plt.axis('equal')
    plt.show()
    print('loss: {}'.format(kmeans.inertia_))
```

```python
init_centres = np.array([[0,3],[0,5]])
do_kmeans(init_centres,X)
```

```python
init_centres = np.array([[3,1],[1,4]])
do_kmeans(init_centres,X)
```
